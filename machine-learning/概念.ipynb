概念定义是在实例集合之上，这个集合表示为$x$($x$为特征值构成的向量)，待学习的概念或者目标函数标记为$c$

$c(x) = 1$ 运动 $c(x) = 0$ 不运动 $c(x)$ 也可以叫 $y$

$x$：表示每一个实例

$X$: 表示每一个样例 所有实例的集合

学习目标 $f: X -> Y$


#### 训练集：用来进行训练，产生模型或者算法的数据集
#### 测试集： 用来专门进行测试已经学习好的模型后者算法的数据集
#### 特征向量： 属性的集合，通常用一个向量来表示，附属于一个实例
#### 标记： 实例类别标记 $c(x)$
#### 正例
#### 反例
