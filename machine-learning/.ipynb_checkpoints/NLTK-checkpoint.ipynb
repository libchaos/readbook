```python
import nltk
```

```python
from nltk.tokenize import PunktSentenceTokenizer
```

```python
train_text = state_union.raw('2005-GWBush.txt')
print(train_text)
```

```python
sample_text = state_union.raw('2006-GWBush.txt')
print(sample_text)
```

```python
custom_sent_tokenizer = PunktSentenceTokenizer(train_text)
```

```python
tokenized = custom_sent_tokenizer.tokenize(sample_text)
```

```python
def process_content():
    try:
        for i in tokenized:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
#             print(tagged)
            chunkGram = r'''Chunk: {<.*>+}
                }<VB.?|IN|DT|TO>+{'''
            chunkParser = nltk.RegexpParser(chunkGram)
            chunked = chunkParser.parse(tagged)
            print(chunked)
    except Exception as e:
        print(str(e))
```

```python
process_content()
```

```python
def process_content():
    try:
        for i in tokenized:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
#             print(tagged)
            nameEnt = nltk.ne_chunk(tagged)
            print(nameEnt)
    except Exception as e:
        print(str(e))
```

```python
process_content()
```

```python
from nltk.corpus import wordnet
```

```python
syns = wordnet.synsets('program')
```

```python
print(syns)
```

```python
print(syns[0].lemmas()[0].name())
```

```python
synonyms = []
antonyms = []
for syn in wordnet.synsets('good'):
    for l in syn.lemmas():
        print('l: ', l)
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())
```

```python
print(set(synonyms))
print(set(antonyms))
```

```python
import random
```

```python
from nltk.corpus import movie_reviews
```

```python
documents = [(list(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)]
```

```python
random.shuffle(documents)
```

```python
print(documents[1])
```

```python
all_words = []
for w in movie_reviews.words():
    all_words.append(w.lower())
```

```python
all_words = nltk.FreqDist(all_words)
```

```python
print(all_words.most_common(15))
```

```python
print(all_words['stupid'])
```

```python
word_features = list(all_words.keys())[:3000]
```

```python
def find_feature(document):
    words = set(document)
    features = {}
    for w in word_features:
        features[w] = (w in words)
    return features
```

```python
print(find_feature(movie_reviews.words('neg/cv000_29416.txt')))
features = [(find_feature(rev), category) for (rev, category) in documents]
```

```python
training_set = features[:1900]
testing_set = features[1900:]
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
print('Naive Algo accuracy: ', (nltk.classify.accuracy(classifier, testing_set)) * 100)
```

```python
classifier.show_most_informative_features()
```

```python
import pickle
```

```python
save_classifier = open('naivebayes.pickle', 'wb')
pickle.dump(classifier, save_classifier)
```
